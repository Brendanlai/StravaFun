```python
import os
import shutil
```

```python
path = os.getcwd()
target = path + "/23_10_2023_download/activities"
activities = os.listdir(target)
```

```python
fit = [x for x in activities if 'fit' in x]

source_folder = target
destination_folder = os.getcwd() + '/23_10_2023_download/fitFiles'

# Ensure that both source and destination folders exist
if not os.path.exists(source_folder):
    print(f"The source folder '{source_folder}' does not exist.")
elif not os.path.exists(destination_folder):
    print(f"The destination folder '{destination_folder}' does not exist.")
else:
    # List all files in the source folder
    # files = os.listdir(source_folder)
    files = fit

    # Move each file from source to destination
    for file in files:
        source_path = os.path.join(source_folder, file)
        destination_path = os.path.join(destination_folder, file)

        # Check if the file is a regular file (not a directory)
        if os.path.isfile(source_path):
            shutil.move(source_path, destination_path)
    
    print("Moved all the files successfully")
```

```python
folder_path = destination_folder

# Ensure that the folder exists
if not os.path.exists(folder_path):
    print(f"The folder '{folder_path}' does not exist.")
else:
    # List all files in the folder
    files = os.listdir(folder_path)
    files = [x for x in files if 'gz' in x] # Delete the GZ files

    # Iterate through the files and delete them
    for file in files:
        file_path = os.path.join(folder_path, file)
        try:
            if os.path.isfile(file_path):
                os.remove(file_path)
                # print(f"Deleted '{file}'.")
        except Exception as e:
            print(f"Error deleting '{file}': {e}")

    print("All files deleted successfully.")
```
